```python
# ============================================================
# Cell-Free Massive MIMO — Sionna RT (Munich) Downlink Analizi
# 16 AP (Tx) - çok kullanıcı (Rx) senaryosu, Ray Tracing tabanlı kanal.
# Bölümler: per-SNR, per-location, per-AP-count, SDMA, FDMA, TDMA.
# ============================================================


#bu bölüm kurulum ve kütüphane eklentilerini içerir. 

import os
# GPU varsa onu kullan, yoksa "" yazilirsa CPU'ya duser; ortam degiskeni ile secilir.
if os.getenv("CUDA_VISIBLE_DEVICES") is None:
    gpu_num = 0  # Use "" to use the CPU
    os.environ["CUDA_VISIBLE_DEVICES"] = f"{gpu_num}"
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# Import Sionna
try:
    import sionna.phy
except ImportError as e:
    import sys
    if 'google.colab' in sys.modules:
        # Install Sionna in Google Colab
        print("Installing Sionna and restarting the runtime. Please run the cell again.")
        os.system("pip install sionna")
        os.kill(os.getpid(), 5)
    else:
        raise e

# Tek GPU kullan ve sadece gerektiği kadar bellek ayır
import tensorflow as tf
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        tf.config.experimental.set_memory_growth(gpus[0], True)
    except RuntimeError as e:
        print(e)
tf.get_logger().setLevel('ERROR')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sionna.phy.mapping import BinarySource, Constellation, Mapper, Demapper
from sionna.phy.ofdm import ResourceGrid, ResourceGridMapper
from sionna.phy.channel import subcarrier_frequencies, cir_to_ofdm_channel
from sionna.phy.utils import ebnodb2no, compute_ber
from sionna.rt import load_scene, Camera, Transmitter, Receiver, PlanarArray, \
                      PathSolver, RadioMapSolver

# --- Pandas görüntüleme ayarları (tabloları kısaltma) ---
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# --- Tekrarlanabilirlik için tohum ---
my_seed = 42
np.random.seed(my_seed)
tf.random.set_seed(my_seed)
sionna.phy.config.seed = my_seed
print(f"Sistem '{my_seed}' tohumu ile kilitlendi. Sonuclar artik tekrarlanabilir!")
no_preview = True  # False yapilirsa render yerine interaktif preview kullanilir








# SISTEM PARAMETRELERI

subcarrier_spacing = 30e3      # Hz
num_tx = 16                    # AP sayisi
num_tx_ant = 1                 # Her AP'de 1 anten
num_rx_ant = 1                 # Her UE'de 1 anten
batch_size_payload = 1000     # Monte Carlo paket sayisi, 1000 üzeri sayılarla CPU kullanıldığı durumlarda simülasyon hata verebilir.
fft_size = 128                 # subcarrier sayisi
num_ofdm_symbols = 14          # bir slottaki OFDM sembol sayisi
num_time_steps = num_ofdm_symbols  # zaman adimi = slottaki sembol sayisi (bilincli esitleme)
num_bits_per_symbol = 2        # QPSK



# SAHNE VE AP YERLESIMI

scene = load_scene(sionna.rt.scene.munich) #kullanılan sahne built-in münich şehridir. 

min_x, max_x = 0, 400
min_y, max_y = 0, 400
ap_height = 10

x_coords = np.linspace(min_x, max_x, 4)
y_coords = np.linspace(min_y, max_y, 4)

ap_positions = []
for x in x_coords:
    for y in y_coords:
        ap_positions.append([x, y, ap_height])
ap_positions = np.array(ap_positions)

# Sahadaki tum tx'ler kendi icinde ozdes, tum rx'ler kendi icinde ozdes.
# Bu yuzden tx ve rx icin tek bir anten dizisi (PlanarArray) tanimlanir.
scene.tx_array = PlanarArray(num_rows=1,
                             num_cols=1,            # yatayda kac anten  burada 1 (her AP tek antenli)
                             vertical_spacing=0.5,  # elemanlar arasi dikey mesafe (yarim dalga boyu)
                             horizontal_spacing=0.5,
                             pattern="dipole",      # dipol: 360 derece omni-yonlu paterndir (yonlu degil)
                             polarization="V")      # dikey polarizasyon. "cross" olsaydi her AP'de anten 2'ye cikardi.

for i in range(num_tx):
    if f"tx-{i}" in scene.transmitters:   # ilk calistirmada hata vermesin
        scene.remove(f"tx-{i}")
    # Tx'lere hiz vektoru verilmedi -> tx kaynakli Doppler kaymasi olmaz.
    # ilk pozisyon (0,0) noktasidir, yani AP0 oraya yerlestirilir.
    tx = Transmitter(name=f"tx-{i}",
                     position=ap_positions[i],
                     display_radius=8.,
                     color=(1, 0, 0))
    scene.add(tx)

# Kus bakisi kamera (AP'lerin merkezine bakar)
bird_cam = Camera(position=[200, 200, 1200], look_at=[200, 200, 0])

max_depth = 3   # bir isinin sahnede yapabilecegi maksimum yansima/etkilesim sayisi




# RADIO MAP

rm_solver = RadioMapSolver()
# cell_size: hesaplama izgarasinin cozunurlugu. 400x400 alan 1x1'lik karelere bolunur,
#            her karedeki guc 10^7 isinla hesaplanir. Secilen UE hangi karede ise
#            o karenin gucu/isinlari o UE icin kullanilir.
# samples_per_tx: her AP'den firlatilan isin sayisi (burada 10 milyon).
rm = rm_solver(scene,
               max_depth=max_depth,
               cell_size=(1., 1.),
               samples_per_tx=10**7)

if no_preview:
    # render: sahnenin yuksek kaliteli statik fotografini ceker.
    # rm_vmin=-110: -110 dBm'den zayif sinyalleri gosterme (modern cihaz siniri).
    # clip_at=12.: goruntuyu yerden 12 m yukseklikte yatay olarak kes.
    scene.render(camera=bird_cam, radio_map=rm, rm_vmin=-110, clip_at=12.)
else:
    # preview: etkilesimli (interaktif) pencere acar.
    scene.preview(radio_map=rm, rm_vmin=-110, clip_at=12.)




# UE POZISYONLARI
# num_pos=1: her 16 AP icin 1 UE pozisyonu ornekleniyor. toplam 16 olası UE pozisyonu elde ederiz. 
min_gain_db = -110  # bu degerin altindaki path gain'leri yok say
max_gain_db = 0
min_dist = 5        # m
max_dist = 400      # m

ue_pos, _ = rm.sample_positions(num_pos=1,  # batch (tx=16) basina 1 konum, toplam 16 UE konumu
                                metric="path_gain",
                                min_val_db=min_gain_db,
                                max_val_db=max_gain_db,
                                min_dist=min_dist,
                                max_dist=max_dist)
# ue_pos sekli: [batch(=AP sayisi), kullanici sayisi(=num_pos), koordinat(=3)]

scene.rx_array = PlanarArray(num_rows=1, num_cols=1, pattern="dipole", polarization="V")

# Daha once kaydedilmis pozisyon varsa tekrarlanabilirlik icin onu kullan
if os.path.exists("ue_positions.npy"):
    positions = np.load("ue_positions.npy")
    print(f"Pozisyonlar yuklendi: {positions.shape}")
else:
    positions = np.reshape(ue_pos, [-1, 3])

for i in range(len(positions)):
    rx_name = f"rx-{i}"
    if rx_name in scene.receivers:
        scene.remove(rx_name)
    rx = Receiver(name=rx_name,
                  position=positions[i],
                  orientation=[0, 0, 0],
                  display_radius=12.,
                  color=(0, 0, 0))
    scene.add(rx)

np.save("ue_positions.npy", positions)
print(f"Pozisyonlar kaydedildi: {positions.shape}")

if no_preview:
    scene.render(camera=bird_cam, radio_map=rm, rm_vmin=-110, rm_vmax=-50, clip_at=12.)
else:
    scene.preview(radio_map=rm, rm_vmin=-110, clip_at=12.)







# KANAL HESABI (CIR -> OFDM frekans cevabi)

p_solver = PathSolver()
# p_solver: sahnedeki tum TX-RX ciftleri arasindaki isinlari (yollari) bulur.
# max_num_paths_per_src: her vericiden cikan en guclu 1000 yolu takip eder;
#                        bu yollarin gecikmesi, acisi ve gucu kaydedilir.
paths = p_solver(scene, max_depth=3, max_num_paths_per_src=10**3)

# CIR: a (kompleks genlik, zaman boyutlu) ve tau (gecikme, zamandan bagimsiz)
# a neden kompleks? Dalga kat ettigi mesafe yuzunden gecikir, gecikme faz
# degisimine yol acar; faz ancak kompleks sayiyla ifade edilir. Ayrica gelen
# yollar bu fazlar yuzunden birbirini sondurebilir/guclendirebilir.
# sampling_frequency: zaman eksenindeki ornekleme adimi (1/sampling_frequency sn).
# num_time_steps: kanalin kac OFDM sembolu boyunca izlenecegi. Bu sahnede hareket
#                 olmadigi icin (Doppler=0) 14 adimin tumu ayni cikar.
a, tau = paths.cir(sampling_frequency=subcarrier_spacing,
                   num_time_steps=num_time_steps,
                   out_type='numpy')

a = np.expand_dims(a, axis=0)
tau = np.expand_dims(tau, axis=0)

# num_tx ve batch boyutlarini yer degistir: her adimda en bastaki boyuttan
# bir UE secebilmek icin batch eksenini one aliyoruz (16 UE -> batch=16).
a = np.transpose(a, [1, 0, 2, 3, 4, 5, 6])
tau = np.transpose(tau, [1, 0, 2, 3])

print("Shape of a:", a.shape)
print("Shape of tau: ", tau.shape)

frequencies = subcarrier_frequencies(fft_size, subcarrier_spacing)
# cir_to_ofdm_channel: zaman-domeni CIR'i (a, tau) frekans cevabina cevirir.
# Her subcarrier f_k icin tum yollari toplar: H(f_k) = sum_i a_i * exp(-j 2pi f_k tau_i).
# Frekans-secici sonumleme (frequency-selective fading) tam burada, tau uzerinden olusur.
# normalize=True: kanali birim ortalama enerjiye ceker.
h_freq = cir_to_ofdm_channel(frequencies, a, tau, normalize=True)
print("Shape of h_freq:", h_freq.shape)

# Doppler=0 oldugu icin 14 zaman adiminin tumu ayni cikar (kontrol amacli yazdiriyoruz)
print(a[0, 0, 0, 0, 0, 0, :])




# ORTAK FONKSIYON VE NESNELER (tum senaryolarda kullanilir)

ebnos_db = np.arange(-5, 15, 2)
constellation = Constellation("qam", num_bits_per_symbol=2)
binary_source = BinarySource()
mapper = Mapper(constellation=constellation)
new_demapper = Demapper(demapping_method="app", constellation=constellation)




# 1) PER-SNR ANALIZI (tek UE, MRT precoding, SNR taramasi)

rg_per_snr = ResourceGrid(num_ofdm_symbols=num_ofdm_symbols,
                          fft_size=fft_size,
                          subcarrier_spacing=30e3,
                          num_tx=1,
                          cyclic_prefix_length=10,
                          num_streams_per_tx=1,
                          pilot_pattern=None,
                          num_guard_carriers=[0, 0],
                          dc_null=False)

rg_mapper_per_snr = ResourceGridMapper(rg_per_snr)
bits_per_snr = binary_source([batch_size_payload, 1, 1,
                              rg_per_snr.num_data_symbols * num_bits_per_symbol])
x = mapper(bits_per_snr)
x_rg_per_snr = rg_mapper_per_snr(x)
x_signal_per_snr = tf.reshape(x_rg_per_snr, [batch_size_payload, 14, 128])

print(f"shape of grid {x_rg_per_snr.shape}")
print(f"Sinyal Boyutu: {x_signal_per_snr.shape}")

loc_idx_per_snr = 1  # h_freq'ten 1 numarali UE'nin 16 AP kanalini sec
# Indeksleme mantigi: tek sayi yazdigin boyutlar "ezilir" ve silinir
# (loc_idx, RX=0, Path=0). Iki nokta ":" koydugun boyutlar korunur.
# Sonuc: [16 AP, 14 zaman, 128 subcarrier]
h_ue_per_snr = tf.cast(h_freq[loc_idx_per_snr, 0, 0, :, 0, :, :], tf.complex64)  # [16 AP, 14, 128]
print(f"Kanal Boyutu: {h_ue_per_snr.shape}")

# MRT precoder: w = conj(h). Her AP, kendi kanalinin esleniginden agirlik uretir;
# boylece havada sinyaller ayni fazda toplanir (maximum ratio transmission).
w_mrt_per_snr = tf.math.conj(h_ue_per_snr)
print(f"w_mrt_per_snr Boyutu: {w_mrt_per_snr.shape}")
# Guc normalizasyonu: w'yu birim vektore cevirir, cikis gucu patlamaz.
w_mrt_per_snr = w_mrt_per_snr / tf.cast(tf.norm(w_mrt_per_snr, axis=0, keepdims=True) + 1e-9, tf.complex64)

# Carpim icin x_signal'e AP (tx) boyutu ekle
x_signal_expanded_per_snr = tf.expand_dims(x_signal_per_snr, axis=1)  # [1000, 1, 14, 128]
x_precoded_per_snr = w_mrt_per_snr * x_signal_expanded_per_snr        # [1000, 16, 14, 128]
print(f"precoded Sinyal Boyutu: {x_precoded_per_snr.shape}")

# Havada birlesme (16 AP uzerinden toplam) -> UE anteninde toplanan sinyal
y_received_per_snr = tf.reduce_sum(h_ue_per_snr * x_precoded_per_snr, axis=1)  # [1000, 14, 128]

# Efektif kanal: precoding dahil edilmis toplam kazanc. Alinan sinyali buna
# bolerek esitleme (equalization) yapariz; boylece kanali sade hale getiririz.
g_eff_per_snr = tf.reduce_sum(h_ue_per_snr * w_mrt_per_snr, axis=0)

ber_list_per_snr = []
for ebno in ebnos_db:
    no_v = ebnodb2no(ebno, num_bits_per_symbol=2, coderate=1.0)

    noise_std = tf.sqrt(no_v / 2)
    noise = tf.complex(tf.random.normal(tf.shape(y_received_per_snr), stddev=noise_std),
                       tf.random.normal(tf.shape(y_received_per_snr), stddev=noise_std))
    y_noisy = y_received_per_snr + noise

    x_hat = y_noisy / tf.cast(g_eff_per_snr + 1e-12, tf.complex64)  # equalization
    x_hat_flat = tf.reshape(x_hat, [batch_size_payload, -1])
    l = new_demapper(x_hat_flat, tf.cast(no_v, tf.float32))

    b_h = tf.cast(l > 0, tf.float32)
    b_f = tf.reshape(b_h, bits_per_snr.shape)
    current_ber = compute_ber(bits_per_snr, b_f).numpy()
    ber_list_per_snr.append(current_ber)
    print(f"EbNo: {ebno} dB | BER: {current_ber:.6f}")

plt.figure(figsize=(10, 6))
plt.semilogy(ebnos_db, ber_list_per_snr, 'r-s', label=f'MRT, UE {loc_idx_per_snr}')
plt.grid(True, which="both")
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.xlim(-5, 12.5)
plt.legend()
plt.show()





# 2) PER-LOCATION ANALIZI (16 UE, sabit SNR'da)

ebno_db_fixed = 5.0
no_val = ebnodb2no(ebno_db=ebno_db_fixed, num_bits_per_symbol=num_bits_per_symbol, coderate=1.0)

rg_per_loc = ResourceGrid(num_ofdm_symbols=14, fft_size=128, subcarrier_spacing=30e3,
                          num_tx=1, num_streams_per_tx=1)
rg_mapper_per_loc = ResourceGridMapper(rg_per_loc)

bits_per_loc = binary_source([batch_size_payload, 1, 1,
                              rg_per_loc.num_data_symbols * num_bits_per_symbol])
x_per_loc = mapper(bits_per_loc)
x_rg_per_loc = rg_mapper_per_loc(x_per_loc)
x_signal_per_loc = tf.reshape(x_rg_per_loc, [batch_size_payload, 14, 128])

ber_list_per_loc = []
print(f"\nEbNo = {ebno_db_fixed} dB icin 16 UE taraniyor (her biri {batch_size_payload} batch)...")

for loc_idx in range(16):
    h_ue_per_loc = tf.cast(h_freq[loc_idx, 0, 0, :, 0, :, :], tf.complex64)  # [16, 14, 128]

    w_mrt_per_loc = tf.math.conj(h_ue_per_loc)
    w_mrt_per_loc = w_mrt_per_loc / tf.cast(tf.norm(w_mrt_per_loc, axis=0, keepdims=True) + 1e-9, tf.complex64)

    x_signal_exp_per_loc = tf.expand_dims(x_signal_per_loc, axis=1)  # [1000, 1, 14, 128]
    x_precoded_per_loc = w_mrt_per_loc * x_signal_exp_per_loc        # [1000, 16, 14, 128]

    y_received_per_loc = tf.reduce_sum(h_ue_per_loc * x_precoded_per_loc, axis=1)  # [1000, 14, 128]

    noise_std_per_loc = tf.sqrt(no_val / 2)
    noise_per_loc = tf.complex(tf.random.normal(tf.shape(y_received_per_loc), stddev=noise_std_per_loc),
                               tf.random.normal(tf.shape(y_received_per_loc), stddev=noise_std_per_loc))
    y_noisy_per_loc = y_received_per_loc + noise_per_loc

    g_eff_per_loc = tf.reduce_sum(h_ue_per_loc * w_mrt_per_loc, axis=0)
    x_hat_per_loc = y_noisy_per_loc / tf.cast(g_eff_per_loc + 1e-12, tf.complex64)

    x_hat_flat_per_loc = tf.reshape(x_hat_per_loc, [batch_size_payload, -1])
    llr = new_demapper(x_hat_flat_per_loc, tf.cast(no_val, tf.float32))

    bits_hat = tf.cast(llr > 0, tf.float32)
    bits_hat_final = tf.reshape(bits_hat, bits_per_loc.shape)

    current_ber = compute_ber(bits_per_loc, bits_hat_final).numpy()
    ber_list_per_loc.append(current_ber)
    print(f"UE {loc_idx:2d} | BER: {current_ber:.6f}")

print("\n--- BATCH ANALIZI TAMAMLANDI ---")

# Her UE icin bagli AP sayisi / en guclu AP
print(f"\n{'UE No':<8} | {'Bagli AP Sayisi':<15} | {'En Guclu AP':<12} | {'Max Guc (Normalize)':<15}")
for u_idx in range(16):
    ap_powers = []
    for a_idx in range(16):
        # h_freq: [UE, TX, RX, AP, Path, Time, Subcarrier]
        h_segment = h_freq[u_idx, 0, 0, a_idx, 0, :, :]
        ap_powers.append(np.sum(np.abs(h_segment)))
    ap_powers = np.array(ap_powers)
    active_ap_count = np.sum(ap_powers > 1.0)
    max_ap = np.argmax(ap_powers)
    max_val = np.max(ap_powers)
    print(f"UE {u_idx:<5} | {active_ap_count:<15} | AP {max_ap:<9} | {max_val:.4f}")


def plot_munich_ber_analysis(rm, ap_positions, ue_pos, ue_ber_results=None):
    """Munich sahnesinde path gain haritasi + AP/UE konumlari (+ istege bagli BER)."""

    def to_numpy(x):
        return x.numpy() if hasattr(x, "numpy") else np.asarray(x)

    pg = to_numpy(rm.path_gain)
    pg_db = 10 * np.log10(np.max(pg, axis=0) + 1e-30)
    centers = to_numpy(rm.cell_centers)

    x_map, y_map = centers[:, :, 1], centers[:, :, 0]
    ap_xy = to_numpy(ap_positions)[:, :2]
    ue_xy = to_numpy(ue_pos).reshape(-1, 3)[:, :2]

    plt.figure(figsize=(12, 10))
    mesh = plt.pcolormesh(x_map, y_map, pg_db, shading="auto", cmap="inferno", vmin=-110, vmax=0)
    cbar = plt.colorbar(mesh)
    cbar.set_label('Path Gain [dB]', fontsize=18, fontweight='bold')

    plt.scatter(ap_xy[:, 1], ap_xy[:, 0], marker="+", s=180, linewidths=2.5,
                color="red", zorder=3, label="Access Points (AP)")
    for i, (x_ap, y_ap) in enumerate(ap_xy):
        plt.text(y_ap, x_ap - 15, f"AP {i}", color="white", fontsize=18,
                 fontweight="bold", zorder=4, ha='center',
                 bbox=dict(facecolor='red', alpha=0.4, edgecolor='none', pad=1))

    for i, (x_ue, y_ue) in enumerate(ue_xy):
        label = f"UE {i}"
        if ue_ber_results is not None:
            label += f"\nBER: {ue_ber_results[i]:.4f}"
        plt.text(y_ue + 15, x_ue + 15, label, color="cyan", fontsize=18,
                 fontweight="bold", zorder=5,
                 bbox=dict(facecolor='black', alpha=0.7, edgecolor='none', pad=1))

    plt.xlabel("Coordinat Y [m]", fontsize=16)
    plt.ylabel("Coordinat X [m]", fontsize=16)

    all_x = np.concatenate([ap_xy[:, 0], ue_xy[:, 0]])
    all_y = np.concatenate([ap_xy[:, 1], ue_xy[:, 1]])
    margin = 80
    plt.xlim(np.max(all_y) + margin, np.min(all_y) - margin)
    plt.ylim(np.min(all_x) - margin, np.max(all_x) + margin)

    plt.grid(True, linestyle='--', alpha=0.3)
    plt.legend(loc='upper right', frameon=True, shadow=True)
    plt.tight_layout()
    plt.show()


# BER hesabi positions uzerinden yapildigi icin haritada da positions kullaniyoruz
plot_munich_ber_analysis(rm, ap_positions, positions, ber_list_per_loc)






# BAGLANTI ANALIZI (UE-AP yol sayisi / fiziksel siddet) -> dosyaya yaz

a_clean = np.squeeze(a)

with open("tum_baglanti_analizi.txt", "w") as f:
    header = f"{'UE No':<6} | {'AP No':<6} | {'Yol Sayisi':<12} | {'Toplam Fiziksel Siddet':<25}"
    f.write(header + "\n")
    f.write("-" * len(header) + "\n")

    for u_idx in range(16):
        found_connection = False
        for ap_idx in range(16):
            try:
                path_data = a_clean[u_idx, ap_idx]
            except Exception:
                path_data = a_clean[ap_idx, u_idx]

            total_power = np.sum(np.abs(path_data))
            if total_power > 1e-20:
                path_magnitudes = np.max(np.abs(path_data), axis=-1)
                active_paths = np.sum(path_magnitudes > 0)
                f.write(f"UE {u_idx:<3} | AP {ap_idx:<3} | {active_paths:<10} | {total_power:.20e}\n")
                found_connection = True
        if not found_connection:
            f.write(f"UE {u_idx:<3} | --- BAGLANTI YOK (SHADOWING) ---\n")
        f.write("-" * len(header) + "\n")

print("Islem tamamlandi! Liste 'tum_baglanti_analizi.txt' dosyasina kaydedildi.")

# Bagli AP'lere mesafe analizi
print(f"\n{'UE No':<8} | {'Bagli AP':<10} | {'En Yakin AP':<12} | {'Min Mesafe (m)':<15} | {'Ort Mesafe (m)':<15}")
print("-" * 75)
for u_idx in range(16):
    ue_position = positions[u_idx]
    ap_powers = np.array([np.sum(np.abs(h_freq[u_idx, 0, 0, a_idx, 0, :, :])) for a_idx in range(16)])
    active_ap_indices = np.where(ap_powers > 1.0)[0]
    if len(active_ap_indices) > 0:
        distances = np.array([np.linalg.norm(ue_position[:2] - ap_positions[ap_idx][:2])
                              for ap_idx in active_ap_indices])
        nearest_ap = active_ap_indices[np.argmin(distances)]
        print(f"UE {u_idx:<5} | {len(active_ap_indices):<10} | AP {nearest_ap:<9} | "
              f"{np.min(distances):<15.1f} | {np.mean(distances):<15.1f}")
    else:
        print(f"UE {u_idx:<5} | 0 (Golge) | ---         | ---            | ---")







# PER-AP-COUNT ANALIZI (sabit SNR'da, AP sayisinin etkisi)

ap_counts = [1, 2, 4, 8, 16]

rg_per_ap = ResourceGrid(num_ofdm_symbols=num_ofdm_symbols, fft_size=fft_size,
                         subcarrier_spacing=30e3, num_tx=1, num_streams_per_tx=1,
                         pilot_pattern=None, num_guard_carriers=[0, 0], dc_null=False)
rg_mapper_per_ap = ResourceGridMapper(rg_per_ap)

bits_per_ap = binary_source([batch_size_payload, 1, 1,
                             rg_per_ap.num_data_symbols * num_bits_per_symbol])
x_per_ap = mapper(bits_per_ap)
x_rg_per_ap = rg_mapper_per_ap(x_per_ap)
x_signal_per_ap = tf.reshape(x_rg_per_ap, [batch_size_payload, 14, 128])

ber_list_per_ap = []
print(f"\n--- EbNo = {ebno_db_fixed} dB sabit calisma noktasinda 16 UE analiz ediliyor ---")

for loc_idx in range(16):
    ue_row = {"UE No": loc_idx}
    for count in ap_counts:
        # En yakin 'count' AP'yi sec
        ue_p = positions[loc_idx]
        distances = np.linalg.norm(ap_positions - ue_p, axis=1)
        best_ap_indices = np.argsort(distances)[:count]

        h_ue_all = tf.cast(h_freq[loc_idx, 0, 0, :, 0, :, :], tf.complex64)
        h_ue_selected = tf.gather(h_ue_all, best_ap_indices, axis=0)

        w_mrt = tf.math.conj(h_ue_selected)
        w_mrt = w_mrt / tf.cast(tf.norm(w_mrt, axis=0, keepdims=True) + 1e-9, tf.complex64)
        g_eff = tf.reduce_sum(h_ue_selected * w_mrt, axis=0)

        # Basitlestirilmis MRT modeli: x * g_eff
        y_received = x_signal_per_ap * tf.cast(g_eff, tf.complex64)
        noise_std = tf.sqrt(no_val / 2)
        noise = tf.complex(tf.random.normal(tf.shape(y_received), stddev=noise_std),
                           tf.random.normal(tf.shape(y_received), stddev=noise_std))
        y_noisy = y_received + noise

        x_hat = y_noisy / tf.cast(g_eff + 1e-12, tf.complex64)
        x_hat_flat = tf.reshape(x_hat, [batch_size_payload, -1])
        l = new_demapper(x_hat_flat, tf.cast(no_val, tf.float32))

        b_h = tf.cast(l > 0, tf.float32)
        b_f = tf.reshape(b_h, bits_per_ap.shape)
        current_ber = compute_ber(bits_per_ap, b_f).numpy()
        ue_row[f"{count} AP"] = round(current_ber, 6)
    ber_list_per_ap.append(ue_row)

df_final = pd.DataFrame(ber_list_per_ap).set_index("UE No")
print("\n" + "=" * 65)
print("   MUNIH SAHNESI: UE BAZLI AP SAYISI PERFORMANSI")
print("=" * 65)
print(df_final.to_string())


# --- Tek UE icin AP sayisi vs SNR taramasi ---
plt.rcParams.update({
    'font.size': 20, 'axes.labelsize': 25, 'axes.titlesize': 25,
    'xtick.labelsize': 20, 'ytick.labelsize': 20, 'legend.fontsize': 22,
    'lines.linewidth': 3, 'lines.markersize': 10
})

target_ue = 11
ebnos_db_sweep = np.arange(-5, 16, 2)
plt.figure(figsize=(12, 8))
colors = ['#d62728', '#ff7f0e', '#2ca02c', '#1f77b4', '#9467bd']
print(f"--- UE {target_ue} icin AP sayisi performans analizi ---")

for i, count in enumerate(ap_counts):
    ber_list = []
    ue_p = positions[target_ue]
    distances = np.linalg.norm(ap_positions - ue_p, axis=1)
    best_ap_indices = np.argsort(distances)[:count]

    h_ue_all = tf.cast(h_freq[target_ue, 0, 0, :, 0, :, :], tf.complex64)
    h_ue_selected = tf.gather(h_ue_all, best_ap_indices, axis=0)

    w_mrt = tf.math.conj(h_ue_selected)
    w_mrt = w_mrt / tf.cast(tf.norm(w_mrt, axis=0, keepdims=True) + 1e-9, tf.complex64)
    g_eff = tf.reduce_sum(h_ue_selected * w_mrt, axis=0)

    for ebno in ebnos_db_sweep:
        no_v = ebnodb2no(ebno, num_bits_per_symbol=2, coderate=1.0)
        y_received = x_signal_per_ap * tf.cast(g_eff, tf.complex64)
        noise_std = tf.sqrt(no_v / 2)
        noise = tf.complex(tf.random.normal(tf.shape(y_received), stddev=noise_std),
                           tf.random.normal(tf.shape(y_received), stddev=noise_std))
        y_noisy = y_received + noise

        x_hat = y_noisy / tf.cast(g_eff + 1e-12, tf.complex64)
        x_hat_flat = tf.reshape(x_hat, [batch_size_payload, -1])
        l = new_demapper(x_hat_flat, tf.cast(no_v, tf.float32))
        b_h = tf.cast(l > 0, tf.float32)
        b_f = tf.reshape(b_h, bits_per_ap.shape)
        ber_list.append(compute_ber(bits_per_ap, b_f).numpy())

    plt.semilogy(ebnos_db_sweep, ber_list, marker='o', label=f'{count} AP', color=colors[i], linewidth=2)
    print(f"AP Count {count} tamamlandi.")

plt.grid(True, which="both", linestyle='--', alpha=0.5)
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.title(f'UE {target_ue}', fontweight='bold')
plt.legend()
plt.ylim([1e-5, 1])
plt.xlim([-5, 15])
plt.show()








# SDMA — Space Division Multiple Access
# 2 UE, ayni zaman-frekans kaynagi, ZF precoding ile uzamsal ayrim.

ue_number = 2
rg_sdma = ResourceGrid(num_ofdm_symbols=num_ofdm_symbols, fft_size=fft_size,
                       subcarrier_spacing=30e3, num_tx=1, num_streams_per_tx=ue_number,
                       pilot_pattern=None, num_guard_carriers=[0, 0], dc_null=False)
rg_mapper_sdma = ResourceGridMapper(rg_sdma)

bits_sdma = binary_source([batch_size_payload, 1, ue_number,
                           rg_sdma.num_data_symbols * num_bits_per_symbol])
x_sdma = mapper(bits_sdma)
x_rg_sdma = rg_mapper_sdma(x_sdma)
x_signal_sdma = tf.expand_dims(tf.reshape(x_rg_sdma, [batch_size_payload, ue_number, 14, 128]), axis=2)
print(f"Uretilen Grid Boyutu: {x_signal_sdma.shape}")

# 2 UE'nin kanali (6 ve 7 indeksli UE'ler)
h_ue_sdma = tf.cast(h_freq[6:6+ue_number, 0, 0, :, 0, :, :], tf.complex64)
H_sdma = tf.cast(h_ue_sdma, tf.complex64)  # [2, 16, 14, 128] (UE, AP, Time, Freq)

# Her subcarrier ve sembol icin diklik (ZF): [Time, Freq, UE, AP]
H_ready_sdma = tf.transpose(H_sdma, [2, 3, 0, 1])
H_hermitian_sdma = tf.linalg.adjoint(H_ready_sdma)

gram_sdma = tf.matmul(H_ready_sdma, H_hermitian_sdma)
gram_inv_sdma = tf.linalg.inv(gram_sdma + tf.eye(ue_number, dtype=tf.complex64) * 1e-9)
W_matrix_sdma = tf.matmul(H_hermitian_sdma, gram_inv_sdma)  # [14, 128, 16, 2]

w_zf_sdma = tf.transpose(W_matrix_sdma, [3, 2, 0, 1])  # [UE, AP, Time, Freq]
print(f"w_zf.shape: {w_zf_sdma.shape}")

# Null-space testi: UE1'den cikanin UE0'da sifir olmasi
nulling_check = tf.reduce_mean(tf.abs(tf.reduce_sum(H_sdma[1] * w_zf_sdma[0], axis=0)))
print(f"Sizinti (Null-Space basarisi): {nulling_check.numpy():.10f}")

# Guc normalizasyonu
norm_factor_sdma = tf.sqrt(tf.reduce_sum(tf.abs(w_zf_sdma)**2, axis=1, keepdims=True))
w_zf_norm_sdma = tf.expand_dims(w_zf_sdma / tf.cast(norm_factor_sdma, tf.complex64), axis=0)

x_precoded_sdma = tf.expand_dims(tf.reduce_sum(w_zf_norm_sdma * x_signal_sdma, axis=1), axis=1)
print(f"x_precoded boyutu: {x_precoded_sdma.shape}")

y_received_sdma = tf.reduce_sum(H_sdma * x_precoded_sdma, axis=2)
print(y_received_sdma.shape)

ber_list_sdma = []
for ebno in ebnos_db:
    no_v = ebnodb2no(ebno, num_bits_per_symbol=2, coderate=1.0)

    # ZF + guc normalizasyonu sonrasi efektif kazanc reel pozitif oldugu icin
    # hard-decision QPSK'da ayrica equalization gerekmez.
    noise_std = tf.sqrt(no_v / 2)
    noise = tf.complex(tf.random.normal(tf.shape(y_received_sdma), stddev=noise_std),
                       tf.random.normal(tf.shape(y_received_sdma), stddev=noise_std))
    y_noisy = y_received_sdma + noise

    x_input = tf.reshape(y_noisy, [batch_size_payload, 1, ue_number, 1792])
    l = new_demapper(x_input, tf.cast(tf.fill([1], no_v), tf.float32))

    b_h = tf.cast(l > 0, tf.float32)
    b_f = tf.reshape(b_h, bits_sdma.shape)
    current_ber = compute_ber(bits_sdma, b_f).numpy()
    ber_list_sdma.append(current_ber)
    print(f"EbNo: {ebno} dB | BER: {current_ber:.6f}")

plt.figure(figsize=(10, 6))
plt.semilogy(ebnos_db, ber_list_sdma, 'b-o', label='2 UE Cell-Free ZF')
plt.grid(True, which="both")
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.title('Multi-User ZF Precoding Performansi')
plt.legend()
plt.show()





# FDMA — Frequency Division Multiple Access
# 2 UE, 16-QAM, MRT precoding, frekans bandi UE'lere bolunur.

num_bits_per_symbol_fdma = 4  # 16-QAM
constellation_fdma = Constellation("qam", num_bits_per_symbol=4)
mapper_fdma = Mapper(constellation=constellation_fdma)
new_demapper_fdma = Demapper(demapping_method="app", constellation=constellation_fdma)

rg_fdma = ResourceGrid(num_ofdm_symbols=num_ofdm_symbols, fft_size=fft_size,
                       subcarrier_spacing=30e3, num_tx=1, num_streams_per_tx=1,
                       pilot_pattern=None, num_guard_carriers=[0, 0], dc_null=False)
rg_mapper_fdma = ResourceGridMapper(rg_fdma)

active_ue_number = 2

bits_fdma = binary_source([batch_size_payload, 1, rg_fdma.num_data_symbols * num_bits_per_symbol_fdma])
x_fdma = mapper_fdma(bits_fdma)
x_fdma = tf.reshape(x_fdma, [batch_size_payload, 1, 1, -1])
x_rg_fdma = rg_mapper_fdma(x_fdma)
h_fdma = h_freq[6:6+active_ue_number, 0, 0, :, 0, :, :]

w_mrt_fdma = tf.math.conj(h_fdma)
norm_factor_fdma = tf.sqrt(tf.reduce_sum(tf.abs(w_mrt_fdma)**2, axis=1, keepdims=True))
w_mrt_fdma = tf.expand_dims(w_mrt_fdma / tf.cast(norm_factor_fdma, tf.complex64), axis=0)

x_fdma = w_mrt_fdma * x_rg_fdma
y_received_fdma = tf.reduce_sum(h_fdma * x_fdma, axis=2)
print(f"bits: {bits_fdma.shape} | x: {x_fdma.shape} | h: {h_fdma.shape} | y: {y_received_fdma.shape}")

ber_list_fdma = []
for ebno in ebnos_db:
    no_v = ebnodb2no(ebno, num_bits_per_symbol=4, coderate=1.0)

    noise = tf.complex(tf.random.normal(tf.shape(y_received_fdma), stddev=tf.sqrt(no_v/2)),
                       tf.random.normal(tf.shape(y_received_fdma), stddev=tf.sqrt(no_v/2)))
    y_noisy = y_received_fdma + noise

    useful_data = np.zeros((batch_size_payload, 14, 128), dtype=np.complex64)
    number_of_subcarrier_per_user = int(fft_size / active_ue_number)

    for u_idx in range(active_ue_number):
        start_f = u_idx * number_of_subcarrier_per_user
        end_f = start_f + number_of_subcarrier_per_user

        y_ue_current = y_noisy[:, u_idx, :, start_f:end_f]   # frekans bandi dilimle
        h_ue_current = h_fdma[u_idx, :, :, start_f:end_f]

        w_mrt = tf.math.conj(h_ue_current)
        w_mrt = w_mrt / tf.cast(tf.norm(w_mrt, axis=0, keepdims=True) + 1e-9, tf.complex64)
        g_eff = tf.reduce_sum(h_ue_current * w_mrt, axis=0)

        y_ue_current = y_ue_current / g_eff  # equalization
        useful_data[:, :, start_f:end_f] = y_ue_current

    x_input = tf.reshape(useful_data, [batch_size_payload, 1, 1, -1])
    l = new_demapper_fdma(x_input, tf.cast(no_v, tf.float32))

    b_h = tf.cast(l > 0, tf.float32)
    b_f = tf.reshape(b_h, bits_fdma.shape)
    current_ber = compute_ber(bits_fdma, b_f).numpy()
    ber_list_fdma.append(current_ber)
    print(f"eb_no {ebno} BER: {current_ber:.6f}")

plt.figure(figsize=(10, 6))
plt.semilogy(ebnos_db, ber_list_fdma, 'b-o', label='2 UE Cell-Free MRT (FDMA)')
plt.ylim([1e-6, 1])
plt.grid(True, which="both")
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.legend()
plt.show()








# TDMA — Time Division Multiple Access
# 2 UE, 16-QAM, MRT precoding, OFDM sembolleri (zaman) UE'lere bolunur.

rg_tdma = ResourceGrid(num_ofdm_symbols=num_ofdm_symbols, fft_size=fft_size,
                       subcarrier_spacing=30e3, num_tx=1, num_streams_per_tx=1,
                       pilot_pattern=None, num_guard_carriers=[0, 0], dc_null=False)
rg_mapper_tdma = ResourceGridMapper(rg_tdma)

bits_tdma = binary_source([batch_size_payload, 1, rg_tdma.num_data_symbols * num_bits_per_symbol_fdma])
x_tdma = mapper_fdma(bits_tdma)
x_tdma = tf.reshape(x_tdma, [batch_size_payload, 1, 1, -1])
x_rg_tdma = rg_mapper_tdma(x_tdma)
h_tdma = h_freq[6:6+active_ue_number, 0, 0, :, 0, :, :]

w_mrt_tdma = tf.math.conj(h_tdma)
norm_factor_tdma = tf.sqrt(tf.reduce_sum(tf.abs(w_mrt_tdma)**2, axis=1, keepdims=True))
w_mrt_tdma = tf.expand_dims(w_mrt_tdma / tf.cast(norm_factor_tdma, tf.complex64), axis=0)

x_tdma = w_mrt_tdma * x_rg_tdma
y_received_tdma = tf.reduce_sum(h_tdma * x_tdma, axis=2)
print(f"bits: {bits_tdma.shape} | x: {x_tdma.shape} | h: {h_tdma.shape} | y: {y_received_tdma.shape}")

ber_list_tdma = []
for ebno in ebnos_db:
    no_v = ebnodb2no(ebno, num_bits_per_symbol=4, coderate=1.0)

    noise = tf.complex(tf.random.normal(tf.shape(y_received_tdma), stddev=tf.sqrt(no_v/2)),
                       tf.random.normal(tf.shape(y_received_tdma), stddev=tf.sqrt(no_v/2)))
    y_noisy = y_received_tdma + noise
    y_noisy = np.squeeze(y_noisy)

    useful_data = np.zeros((batch_size_payload, 14, 128), dtype=np.complex64)
    symbol_per_ue = int(num_ofdm_symbols / active_ue_number)

    for u_idx in range(active_ue_number):
        start_f = u_idx * symbol_per_ue
        end_f = start_f + symbol_per_ue

        y_ue_current = y_noisy[:, u_idx, start_f:end_f, :]   # zaman (sembol) dilimle
        h_ue_current = h_tdma[u_idx, :, start_f:end_f, :]

        w_mrt = tf.math.conj(h_ue_current)
        w_mrt = w_mrt / tf.cast(tf.norm(w_mrt, axis=0, keepdims=True) + 1e-9, tf.complex64)
        g_eff = tf.reduce_sum(h_ue_current * w_mrt, axis=0)

        y_ue_current = y_ue_current / g_eff  # equalization
        useful_data[:, start_f:end_f, :] = y_ue_current

    x_input = tf.reshape(useful_data, [batch_size_payload, 1, 1, -1])
    l = new_demapper_fdma(x_input, tf.cast(no_v, tf.float32))

    b_h = tf.cast(l > 0, tf.float32)
    b_f = tf.reshape(b_h, bits_tdma.shape)
    current_ber = compute_ber(bits_tdma, b_f).numpy()
    ber_list_tdma.append(current_ber)

plt.figure(figsize=(10, 6))
plt.semilogy(ebnos_db, ber_list_tdma, 'b-o', label='2 UE Cell-Free MRT (TDMA)')
plt.grid(True, which="both")
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.legend()
plt.show()





# KARSILASTIRMA: TDMA vs FDMA vs SDMA

plt.rcParams.update({
    'font.size': 20, 'axes.labelsize': 25, 'axes.titlesize': 25,
    'xtick.labelsize': 20, 'ytick.labelsize': 20, 'legend.fontsize': 22,
    'lines.linewidth': 3, 'lines.markersize': 10
})

plt.figure(figsize=(12, 8))
plt.semilogy(ebnos_db, ber_list_tdma, 'g-s', label='TDMA (2 UE, MRT)')
plt.semilogy(ebnos_db, ber_list_fdma, 'r-^', label='FDMA (2 UE, MRT)')
plt.semilogy(ebnos_db, ber_list_sdma, 'b-o', label='SDMA (2 UE, ZF)')
plt.grid(True, which="both", linestyle='--', alpha=0.5)
plt.xlabel('Eb/No (dB)')
plt.ylabel('Bit Error Rate (BER)')
plt.legend(loc='lower left')
plt.ylim([1e-5, 1])
plt.xlim([-5, 15])
plt.tight_layout()
plt.show()
```
